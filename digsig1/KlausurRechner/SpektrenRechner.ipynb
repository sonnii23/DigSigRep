```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Ordnung N
N      = 7

#Gleichanteil
gleich = 7

#abtastrate
fs     = 8000

#basisfreq
k     = 2

#Grundfrequenz
f0     = k/N*fs

#Zeitfunktion

x_n    = np.cos(2*np.pi*k/N*np.arange(0,N)) + gleich/N

#prints
print("Grundfrequenz: ",f0)

#plot

plt.stem(x_n)
```

```python
plt.stem(np.fft.fft(x_n))
```
