```python
import numpy as np
import matplotlib.pyplot as plt

# Letzter X-Achsenwert +1 (Ordnung der DFT)
N = 7
# Gleichanteil Ablesen bei x[0]
gleichanteil = 7
# Abtastrate aus der Aufgabenstellung
fs= 8000
# k = laufweite bis zum n√§chsten Ausschlag
k = 2

# Grundfrequenz berechnen
grundfrequenz = k * fs / N

print("Grundfrequenz:")
print(grundfrequenz)

# Zeitfunktion x[n] immer cos Funktion
x_n = np.cos(2 * np.pi / N * k * np.arange(N)) + gleichanteil / N

print("Werte von x[n]")
print(x_n)
print("Zeitfunktion x[n]=cos(2pi/", N, "", k , "*n) +",  gleichanteil, "/",N)

plt.stem(np.arange(N), x_n)

```
