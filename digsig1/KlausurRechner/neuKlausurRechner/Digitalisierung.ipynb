```python
import numpy as np
import matplotlib.pyplot as plt

# Frequency in Hz
f = 1000
# Phase in radians
phase = 0
# Amplitude in V
amplitude = 1

# Time array for two cycles
t = np.linspace(0, 2 / f)

# Generate the sinusoidal signal
signal = amplitude * np.sin(2 * np.pi * f * t + phase)

# Calculate the sampling frequency
werte = 5
fs = f * werte

print("Frequenz: ", f)
print("Abtastrate: ", fs)

if(fs < 2*f):
    print("Abtastrate ist zu klein")
else:
    print("Abtastrate ist ausreichend und das Abtasttheorem ist erf√ºllt")

if(f < 10):
    # Plot the signal
    plt.plot(t, signal)
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude (V)')
    plt.title('Kontinuierliches Sinus Signal')
    plt.grid(True)
    plt.show()
else:
    # Plot the signal
    plt.plot(t * 1000, signal)
    plt.xlabel('Time (ms)')
    plt.ylabel('Amplitude (V)')
    plt.title('Kontinuierliches Sinus Signal')
    plt.grid(True)
    plt.show()
```
