Die Grundfrequenz f0 bei einer Abtastrate fs = 8000 Abtastwerte/s ergibt sich aus f0=k/N*fs

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k   = 5
N   = 16
X_0 = 8
fs  = 8000
f0  =  k/N*fs

p   = 0.05
pn  = np.arange(0, N, p)
print(f0)
```

```python
x_n = np.cos(2*np.pi*f0*np.arange(N)/fs)

plt.stem(x_n)
plt.plot(x_n)
```

```python
def fourierGanz(N, k, n_N, gleichanteil,teiler=1):
    xn = gleichanteil/N + np.cos((2*np.pi)/N * k * n_N)
    #xn = 1/2 + np.cos(2*np.pi/3 *n_N[:4])
    p = 0.05
    pn = np.arange(0, N,p)

    kp = int((k-1)*1/p)+1
    pxn = gleichanteil/N + np.cos((2*np.pi)/N * k * pn)
    plt.subplot(2, 1, 1)
    plt.stem(n_N, xn, 'r', markerfmt='ro')
    plt.plot(pn, pxn, 'b')
    print(gleichanteil/N, "+cos(2pi/", N, "*", k, "*n)")
    from numpy.fft import fft, fftshift
    fxn = gleichanteil/N + np.cos((2*np.pi)/N * k * n_N[0:N:teiler])
    ff= np.abs((fft(fxn,int(N/teiler))))
    plt.subplot(2, 1, 2)
    plt.stem(n_N[:int(N/teiler)], ff, 'b', markerfmt='bo')
```

```python
fourierGanz(N, k, np.arange(N), 0)
```
