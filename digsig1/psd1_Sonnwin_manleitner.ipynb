## Aufgabe 4
#### Skizziere die Folgen x1[n] = σ(n + 3) und x2[n] = δ(n − 2) für den Wertebereich n = −5...5.

```python
#Bibliotheken importieren
import numpy as np  
import matplotlib.pyplot as plt 
#variablen setzen
n = np.arange(-5,5+1)
sprung = np.heaviside(n+3,1) 
#dirac funktion erstellen
def dirac(n,y):
    return (np.array([1 if i == y else 0 for i in n]))+y
#ausgabe plotten
```

## Aufgabe 5
#### Plottet mit hilfe von matplotlib die vorherigen Folgen.


```python
#plt.stem(n,sprung)
#plt.grid(True)
fig, axs = plt.subplots(1,2, figsize=(15, 3))

#verschiedene windows
axs[0].stem(n, dirac(n-2,0),'green')
axs[1].stem(n, sprung ,'blue')
```

## Aufgabe 6
#### Skizziere die Folgen x3[n] = sin(2πn/6) und x4[n] = cos(n/3) für n = 0...10.

```python
#variablen setzen
n2 = np.arange(0,10+1)
sinus = np.sin(2*np.pi*n2/6)
cosinus = np.cos(n2/3)
#plotten
fig, axs = plt.subplots(1,2, figsize=(15, 3))
axs[0].stem(n2, sinus,'green')
axs[1].stem(n2, cosinus ,'blue')
```
