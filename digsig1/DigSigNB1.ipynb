```python
import matplotlib.pyplot as plt
import numpy as np
```

## Variablen setzen:

```python
n_start = -5
n_stop = 10
M = n_stop - n_start
n_M = np.arange(n_start, n_stop)
N = 5
Omega = 2*np.pi/N
x_M = np.cos(Omega*n_M)

def disCosFol(): 
    plt.stem(n_M, x_M)
    plt.xlabel("n")
    plt.ylabel("x")
    plt.title("diskrete cos-folge")
```

## Funktionsaufruf und Plot:
