```python
import numpy as np 
import matplotlib.pyplot as plt 
```

## Funktion definieren 
komplexe Exponentialfolge e^j2pi/N*k

```python
N_dft = 0
n_N = np.arange(N_dft)
k = 1
x_N = np.exp(1j*(2*np.pi/k*N_dft))
```

## plot

```python
plt.plot(x_N.real, x_N.imag, 'bo')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Complex Variable x_N')
plt.show()
```
